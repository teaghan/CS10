## Random numbers

In some of these programs, you will need to generate a random number. 

You can generate a
pseudorandom number in Python using the `random` module and the `randint()` function. (You’ll
learn about modules and functions soon!) Pseudorandom numbers are not truly random
because they are generated by a deterministic computation, but just by looking at the numbers
it is all but impossible to distinguish them from random. 

Below is some example code:

```python
# Program to generate random numbers

# Import the random module
import random

# Generate a pseudorandom number between 0 and 9
x = random.randint(0,9)
print(x)
```
